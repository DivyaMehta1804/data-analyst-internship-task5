```python
import pandas as pd
df = pd.read_csv('train.csv')

# describe()
print("\n--- Describe ---")
print(df.describe(include='all')) 

# info()
print("\n--- Info ---")
print(df.info())

# value_counts() on each column
print("\n--- Value Counts for Each Column ---")
for column in df.columns:
    print(f"\nValue Counts for '{column}':")
    print(df[column].value_counts())


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the data
df = pd.read_csv('train.csv')  # use correct path if needed

# Visualize with sns.pairplot()
print("\n--- Pairplot ---")
sns.pairplot(df)
plt.show()

# Visualize correlation matrix with sns.heatmap()
print("\n--- Heatmap ---")
plt.figure(figsize=(10,8))
correlation_matrix = df.corr(numeric_only=True)  # Only numeric columns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('train.csv')  

# Plot Histograms
print("\n--- Histograms ---")
df.hist(bins=30, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Features')
plt.show()

# Plot Boxplots for each numeric column
print("\n--- Boxplots ---")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 4, i)  # adjust 3x4 depending on number of columns
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# Plot Scatterplots (example between two numeric columns)
print("\n--- Scatterplot Example ---")
if len(numeric_cols) >= 2:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[numeric_cols[0]], y=df[numeric_cols[1]], hue=df[numeric_cols[1]])
    plt.title(f'Scatterplot: {numeric_cols[0]} vs {numeric_cols[1]}')
    plt.xlabel(numeric_cols[0])
    plt.ylabel(numeric_cols[1])
    plt.show()
else:
    print("Not enough numeric columns for scatterplot.")
```
